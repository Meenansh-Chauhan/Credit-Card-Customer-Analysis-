# Section B: Transaction Analysis

## Purpose

Analyze customer spending behaviour through transaction data.

Objectives:

- Calculate key transaction KPIs
- Identify high-performing spending categories
- Analyze monthly revenue trends
- Understand transaction value distribution
- Identify top spending customers

```python
# =============================================================================
# Section B: Transaction Analysis
#
# Purpose:
# Analyze customer spending behaviour through transaction data and identify
# revenue trends, spending patterns, and top-performing categories.
#
# This section includes:
# 1. Data preparation
# 2. Transaction KPIs
# 3. Revenue by category
# 4. Monthly revenue trend
# 5. Monthly transaction volume
# 6. Transaction value distribution
# 7. Top spending customers
# =============================================================================

# -----------------------------
# Data Preparation
# -----------------------------

transactions["date"] = pd.to_datetime(transactions["date"])

customers.rename(
    columns={"propert_owner": "property_owner"},
    inplace=True
)

```

```python
# -----------------------------
# Transaction KPI Summary
# -----------------------------

transaction_summary = {

    "Total Transactions": len(transactions),

    "Total Revenue": transactions["amount"].sum(),

    "Average Transaction": transactions["amount"].mean(),

    "Minimum Transaction": transactions["amount"].min(),

    "Maximum Transaction": transactions["amount"].max()

}

print("="*60)
print("TRANSACTION SUMMARY")
print("="*60)

for key, value in transaction_summary.items():

    if isinstance(value, float):
        print(f"{key:<25}: ₹{value:,.2f}")
    else:
        print(f"{key:<25}: {value:,}")

```

```python
# -----------------------------
# Revenue by Category
# -----------------------------

category_summary = (

    transactions

    .groupby("category")

    .agg(

        Total_Transactions=("id","count"),

        Total_Revenue=("amount","sum"),

        Average_Transaction=("amount","mean")

    )

    .sort_values("Total_Revenue", ascending=False)

)

display(category_summary)

plt.figure(figsize=(8,5))

plt.bar(category_summary.index,
        category_summary["Total_Revenue"])

plt.title("Revenue by Spending Category")

plt.xlabel("Category")

plt.ylabel("Revenue")

plt.xticks(rotation=30)

plt.show()

```

```python
# -----------------------------
# Monthly Revenue
# -----------------------------

monthly_summary = (

    transactions

    .groupby("month")

    .agg(

        Total_Transactions=("id","count"),

        Total_Revenue=("amount","sum")

    )

)

display(monthly_summary)

plt.figure(figsize=(9,5))

plt.plot(monthly_summary.index,
         monthly_summary["Total_Revenue"],
         marker="o")

plt.title("Monthly Revenue Trend")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.grid(True)

plt.show()

```

```python
# -----------------------------
# Monthly Transaction Volume
# -----------------------------

plt.figure(figsize=(9,5))

plt.bar(monthly_summary.index,
        monthly_summary["Total_Transactions"])

plt.title("Monthly Transaction Volume")

plt.xlabel("Month")

plt.ylabel("Transactions")

plt.show()

```

```python
# -----------------------------
# Transaction Value Distribution
# -----------------------------

plt.figure(figsize=(8,5))

plt.hist(transactions["amount"], bins=20)

plt.title("Transaction Amount Distribution")

plt.xlabel("Transaction Amount")

plt.ylabel("Frequency")

plt.show()

```

```python
# -----------------------------
# Top 10 Spending Customers
# -----------------------------

top_customers = (

    transactions

    .groupby("customer_id")

    .agg(

        Total_Spending=("amount","sum"),

        Total_Transactions=("id","count")

    )

    .sort_values("Total_Spending", ascending=False)

    .head(10)

)

display(top_customers)

plt.figure(figsize=(10,5))

plt.bar(
    top_customers.index.astype(str),
    top_customers["Total_Spending"]
)

plt.title("Top 10 Spending Customers")

plt.xlabel("Customer ID")

plt.ylabel("Total Spending")

plt.xticks(rotation=45)

plt.show()
```

# Section C: Customer Spending Analysis

## Purpose

Combine customer demographics with transaction data to understand which customer groups contribute the highest spending.

Objectives:

- Merge customer and transaction datasets
- Analyze spending by age group
- Analyze spending by gender
- Analyze spending by income type
- Analyze spending by education
- Analyze spending by occupation
- Analyze spending by housing type

```python
# =============================================================================
# Section C: Customer Spending Analysis
#
# Purpose:
# Combine customer demographics with transaction data to identify
# which customer groups contribute the highest spending.
#
# This section includes:
# 1. Merge datasets
# 2. Age group analysis
# 3. Gender analysis
# 4. Income type analysis
# 5. Education analysis
# 6. Occupation analysis
# 7. Housing type analysis
# =============================================================================

# -----------------------------
# Merge Customer & Transaction Data
# -----------------------------

customer_transactions = transactions.merge(
    customers,
    left_on="customer_id",
    right_on="id",
    how="left"
)

print("=" * 60)
print("MERGED DATASET SHAPE")
print("=" * 60)

print(customer_transactions.shape)

display(customer_transactions.head())

```

```python
# -----------------------------
# Create Age Groups
# -----------------------------

customer_transactions["age_group"] = pd.cut(
    customer_transactions["age"],
    bins=[18,25,35,45,55,70],
    labels=[
        "18-25",
        "26-35",
        "36-45",
        "46-55",
        "56+"
    ]
)

```

```python
# -----------------------------
# Revenue by Age Group
# -----------------------------

age_summary = (

    customer_transactions

    .groupby("age_group", observed=False)

    .agg(

        Total_Revenue=("amount","sum"),

        Total_Transactions=("amount","count"),

        Average_Spending=("amount","mean")

    )

)

display(age_summary)

plt.figure(figsize=(8,5))

plt.bar(
    age_summary.index.astype(str),
    age_summary["Total_Revenue"]
)

plt.title("Revenue by Age Group")

plt.xlabel("Age Group")

plt.ylabel("Revenue")

plt.tight_layout()

plt.savefig("../visuals/revenue_by_age_group.png",
            dpi=300,
            bbox_inches="tight")

plt.show()

```

```python
# -----------------------------
# Revenue by Gender
# -----------------------------

gender_summary = (

    customer_transactions

    .groupby("gender")

    .agg(

        Total_Revenue=("amount","sum"),

        Average_Spending=("amount","mean"),

        Total_Transactions=("amount","count")

    )

    .sort_values("Total_Revenue", ascending=False)

)

display(gender_summary)

plt.figure(figsize=(6,5))

plt.bar(
    gender_summary.index,
    gender_summary["Total_Revenue"]
)

plt.title("Revenue by Gender")

plt.tight_layout()

plt.savefig("../visuals/revenue_by_gender.png",
            dpi=300,
            bbox_inches="tight")

plt.show()

```

```python
# -----------------------------
# Revenue by Income Type
# -----------------------------

income_summary = (

    customer_transactions

    .groupby("type_income")

    .agg(

        Total_Revenue=("amount","sum"),

        Average_Spending=("amount","mean")

    )

    .sort_values("Total_Revenue", ascending=False)

)

display(income_summary)

plt.figure(figsize=(8,5))

plt.bar(
    income_summary.index,
    income_summary["Total_Revenue"]
)

plt.xticks(rotation=30)

plt.title("Revenue by Income Type")

plt.tight_layout()

plt.savefig("../visuals/revenue_by_income_type.png",
            dpi=300,
            bbox_inches="tight")

plt.show()

```

```python
# -----------------------------
# Revenue by Education
# -----------------------------

education_summary = (

    customer_transactions

    .groupby("education")

    .agg(

        Total_Revenue=("amount","sum")

    )

    .sort_values("Total_Revenue", ascending=False)

)

display(education_summary)

plt.figure(figsize=(9,5))

plt.bar(
    education_summary.index,
    education_summary["Total_Revenue"]
)

plt.xticks(rotation=30)

plt.title("Revenue by Education")

plt.tight_layout()

plt.savefig("../visuals/revenue_by_education.png",
            dpi=300,
            bbox_inches="tight")

plt.show()

```

```python
# -----------------------------
# Revenue by Occupation
# -----------------------------

occupation_summary = (

    customer_transactions

    .groupby("type_occupation")

    .agg(

        Total_Revenue=("amount","sum")

    )

    .sort_values("Total_Revenue", ascending=False)

    .head(10)

)

display(occupation_summary)

plt.figure(figsize=(10,5))

plt.bar(
    occupation_summary.index,
    occupation_summary["Total_Revenue"]
)

plt.xticks(rotation=45)

plt.title("Top Occupations by Revenue")

plt.tight_layout()

plt.savefig("../visuals/revenue_by_occupation.png",
            dpi=300,
            bbox_inches="tight")

plt.show()

```

```python
# -----------------------------
# Revenue by Housing Type
# -----------------------------

housing_summary = (

    customer_transactions

    .groupby("housing_type")

    .agg(

        Total_Revenue=("amount","sum")

    )

    .sort_values("Total_Revenue", ascending=False)

)

display(housing_summary)

plt.figure(figsize=(9,5))

plt.bar(
    housing_summary.index,
    housing_summary["Total_Revenue"]
)

plt.xticks(rotation=30)

plt.title("Revenue by Housing Type")

plt.tight_layout()

plt.savefig("../visuals/revenue_by_housing_type.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

# Section D: Customer Segmentation

## Purpose

Create a customer-level analytical dataset by aggregating transaction behaviour and segment customers based on spending patterns.

Objectives:

- Create customer summary dataset
- Analyze spending distribution
- Segment customers
- Compare customer segments
- Export final analytical dataset for Tableau

```python
# =============================================================================
# Section D: Customer Segmentation
#
# Purpose:
# Build a customer-level analytical dataset and segment customers based on
# their spending behaviour.
#
# This section includes:
# 1. Customer Summary Dataset
# 2. Spending Distribution
# 3. Customer Segmentation
# 4. Segment Comparison
# =============================================================================

# -----------------------------
# Customer Summary Dataset
# -----------------------------

customer_summary = (

    transactions

    .groupby("customer_id")

    .agg(

        Total_Transactions=("id","count"),

        Total_Spending=("amount","sum"),

        Average_Transaction=("amount","mean"),

        First_Transaction=("date","min"),

        Last_Transaction=("date","max")

    )

    .reset_index()

)

display(customer_summary.head())


```

```python
# -----------------------------
# Spending Distribution
# -----------------------------

print("="*60)
print("CUSTOMER SPENDING STATISTICS")
print("="*60)

display(customer_summary["Total_Spending"].describe())


plt.figure(figsize=(8,5))

plt.hist(
    customer_summary["Total_Spending"],
    bins=20
)

plt.title("Customer Spending Distribution")

plt.xlabel("Total Spending")

plt.ylabel("Customers")

plt.tight_layout()

plt.savefig(
    "../visuals/customer_spending_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()

```

```python

# -----------------------------
# Spending Percentiles
# -----------------------------

print("="*60)
print("SPENDING PERCENTILES")
print("="*60)

display(

customer_summary["Total_Spending"].quantile(

[0.25,0.50,0.75,0.90,0.95]

)

)
```

```python
# -----------------------------
# Customer Segmentation
# -----------------------------

q1 = customer_summary["Total_Spending"].quantile(0.25)
q3 = customer_summary["Total_Spending"].quantile(0.75)

customer_summary["Spending_Segment"] = np.where(
    customer_summary["Total_Spending"] <= q1,
    "Low Spending",
    np.where(
        customer_summary["Total_Spending"] <= q3,
        "Medium Spending",
        "High Spending"
    )
)

display(customer_summary.head())
```

```python
# -----------------------------
# Segment Summary
# -----------------------------

segment_summary = (

    customer_summary

    .groupby("Spending_Segment")

    .agg(

        Total_Customers=("customer_id","count"),

        Total_Revenue=("Total_Spending","sum"),

        Average_Customer_Spending=("Total_Spending","mean"),

        Average_Transactions=("Total_Transactions","mean")

    )

)

display(segment_summary)
```

```python
plt.figure(figsize=(7,5))

plt.bar(
    segment_summary.index,
    segment_summary["Total_Customers"]
)

plt.title("Customer Distribution by Spending Segment")

plt.xlabel("Segment")

plt.ylabel("Customers")

plt.tight_layout()

plt.savefig(
    "../visuals/customer_segment_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.bar(
    segment_summary.index,
    segment_summary["Total_Revenue"]
)

plt.title("Revenue by Spending Segment")

plt.xlabel("Segment")

plt.ylabel("Revenue")

plt.tight_layout()

plt.savefig(
    "../visuals/revenue_by_segment.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.bar(
    segment_summary.index,
    segment_summary["Average_Customer_Spending"]
)

plt.title("Average Customer Spending")

plt.xlabel("Segment")

plt.ylabel("Average Spending")

plt.tight_layout()

plt.savefig(
    "../visuals/average_spending_by_segment.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.bar(
    segment_summary.index,
    segment_summary["Average_Transactions"]
)

plt.title("Average Transactions by Segment")

plt.xlabel("Segment")

plt.ylabel("Transactions")

plt.tight_layout()

plt.savefig(
    "../visuals/average_transactions_by_segment.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
